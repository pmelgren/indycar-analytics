```python
import pickle
import pandas as pd

with open('../iters_0-1999.pickle','rb') as f:
    iters = pickle.load(f)

with open('../iters_0-2999_run2.pickle','rb') as f:
    iters_r2 = pickle.load(f)

dfs = []
for i,df in enumerate(iters+iters_r2):
    df['Win'] = df.Pos.apply(lambda x: 1 if x==1 else 0)
    df['DNF'] = (1-df.Finished)
    df['Pole'] = df.SP.apply(lambda x: 1 if x==1 else 0)
    df['Indy500Winner'] = ((df.RaceInSeason ==7) & (df.Pos ==1)).astype(int)
    df_agg = df.groupby('Driver').agg(
        Pts = ('Pts','sum'),
        AvgSP = ('SP','mean'),
        FinishPct = ('Finished','mean'),
        AvgPos = ('Pos','mean'),
        Wins = ('Win','sum'),
        Indy500Winner = ('Indy500Winner','max'),
        DNFs = ('DNF','sum'),
        Poles = ('Pole','sum')
    ).reset_index()
    df_agg['Rank'] = df_agg['Pts'].rank(ascending=False)
    df_agg['iter'] = str(i)
    df_agg['Champion'] = (df_agg['Rank'] == 1).astype(int)
    dfs.append(df_agg)

dfa = pd.concat(dfs)    
```

```python
dfa.groupby('Driver').agg(
        Avg_Pts = ('Pts','mean'),
        FinishRate = ('FinishPct','mean'),
        AvgSP = ('AvgSP','mean'),
        AvgFP = ('AvgPos','mean'),
        AvgWins = ('Wins','mean'),
        AvgPoles = ('Poles','mean'),
        AvgStandings = ('Rank','mean'),
        Championships = ('Champion','sum'),
        Indy500Wins = ('Indy500Winner','sum'),
        Best = ('Rank','min'),
        Worst = ('Rank','max'),
        
).sort_values('AvgStandings')
        
```

```python
with open('../iters_0-2499_with_comps.pickle','rb') as f:
    iters_wc = pickle.load(f)

with open('../iters_0-2499_with_comps_run2.pickle','rb') as f:
    iters_wcr2 = pickle.load(f)

dfs2 = []
for i,df in enumerate(iters_wc+iters_wcr2):
    df['Win'] = df.Pos.apply(lambda x: 1 if x==1 else 0)
    df['DNF'] = (1-df.Finished)
    df['Pole'] = df.SP.apply(lambda x: 1 if x==1 else 0)
    df['Indy500Winner'] = ((df.RaceInSeason ==7) & (df.Pos ==1)).astype(int)
    df_agg = df.groupby('Driver').agg(
        Pts = ('Pts','sum'),
        AvgSP = ('SP','mean'),
        FinishPct = ('Finished','mean'),
        AvgPos = ('Pos','mean'),
        Wins = ('Win','sum'),
        Indy500Winner = ('Indy500Winner','max'),
        DNFs = ('DNF','sum'),
        Poles = ('Pole','sum')
    ).reset_index()
    df_agg['Rank'] = df_agg['Pts'].rank(ascending=False)
    df_agg['iter'] = f'{i}_wc'
    df_agg['Champion'] = (df_agg['Rank'] == 1).astype(int)
    dfs2.append(df_agg)

dfa2 = pd.concat(dfs2)    
```

```python
dfa2.groupby('Driver').agg(
        Avg_Pts = ('Pts','mean'),
        FinishRate = ('FinishPct','mean'),
        AvgSP = ('AvgSP','mean'),
        AvgFP = ('AvgPos','mean'),
        AvgWins = ('Wins','mean'),
        AvgPoles = ('Poles','mean'),
        AvgStandings = ('Rank','mean'),
        Championships = ('Champion','sum'),
        Indy500Wins = ('Indy500Winner','sum'),
        Best = ('Rank','min'),
        Worst = ('Rank','max')
).sort_values('AvgStandings')
        
```

```python
df_all = pd.concat([dfa,dfa2])
df_all['FinishPct2'] = df_all['FinishPct']*100
print(df_all.groupby('Driver').agg(
        Avg_Pts = ('Pts','mean'),
        FinishRate = ('FinishPct2','mean'),
        AvgSP = ('AvgSP','mean'),
        AvgFP = ('AvgPos','mean'),
        AvgWins = ('Wins','mean'),
        AvgPoles = ('Poles','mean'),
).sort_values('Avg_Pts',ascending=False).round(1).to_markdown())
print(df_all.groupby('Driver').agg(
        AvgStandings = ('Rank','mean'),
        Championships = ('Champion','sum'),
        Indy500Wins = ('Indy500Winner','sum'),
        Best = ('Rank','min'),
        Worst = ('Rank','max')
).sort_values('AvgStandings').round(1).to_markdown())
```
