```python
import requests

event_id = '6304'
url = f'https://www.indycar.com/Services/IndyStats.svc/EventsSessionDetails?id={event_id}'

response = requests.get(url)
print(response)
```

```python
d = response.json()
d.keys()
```

```python
for k in d.keys():
    if k == 'records':
        continue
    print(k, d[k])
```

```python
import pandas as pd
df = pd.DataFrame(d['records'])
df.sample(3).T
```
